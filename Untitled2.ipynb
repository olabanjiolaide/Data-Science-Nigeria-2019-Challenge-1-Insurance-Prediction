```
train=dftrain
test=dftest.drop('Geo_Code',axis=1)
```

```
x = train.drop(['Claim','Geo_Code'],axis=1)
y = train.Claim
```

```
cate_features_index = np.where(x.dtypes != float)[0]
# catboost feature indexing
```

```
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.90,random_state=123)
# using sklearn train test split
```

```
model = CatBoostClassifier(iterations=15000,depth=11, learning_rate=0.1,eval_metric='AUC',
                bagging_temperature=1,
                one_hot_max_size=11,
                use_best_model=True,l2_leaf_reg=7.0990621030705965, random_seed=123)
```

```
model.fit(x_train,y_train,cat_features=cate_features_index,eval_set=(x_test,y_test),plot=True)
# making the prediction
```

```
print('this is the optimal  Area Under the ROC curve AUC is :{:.6f}'.format(roc_auc_score(y_test,model.predict(x_test))))
```

```
pred = model.predict(test)
pred = pred.astype(np.int)
submission = pd.DataFrame({'Customer Id':test['Customer Id'],'Claim':pred})
# creating a dataframe from the prediction
```
